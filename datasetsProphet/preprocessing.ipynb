To split the CSV file into training, validation and test call the following cell

```python
import csv
from datetime import datetime


def sort_csv_chronologically(input_csv):
    # Read the CSV data
    with open(input_csv, mode="r") as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header
        data = list(reader)

    # Sort the data by date and time
    data.sort(
        key=lambda row: datetime.strptime(f"{row[0]} {row[1]}", "%Y-%m-%d %H:%M:%S")
    )

    # Write the sorted data back to a new CSV file
    with open(input_csv, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(header)  # Write the header
        writer.writerows(data)
```
